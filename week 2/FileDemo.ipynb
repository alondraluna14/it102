# Variable Assignments

* `=` is the assignment operator.
* Define Python symbols to store values.
* Python names are:
  * Case-sensitive.
  * Consist of alphanumeric characters, `a`-`z`, `A`-`Z`, `0`-`9` and `_`.
  * Must start with a letter or `_`.
    - Variable names starts with lower-case letter.
    - Class names starts with a capital letter.
    - Names starting with `_` is **hidden**. ( variables or functions starting with an underscore are typically considered private. This means they are meant for internal use within a class or module and are not intended to be accessed directly from outside.)
    - *Dunder* (double under): names surrounded by `_ _` are typically created by Python that refer to special methods or attributes
  * No length limitations.

### Operator Precedence

- #### Operations higher up in table has higher precedence.
- #### Operators of the same precedence (in the same row) *associates* left-to-right;

| Operator | Description |
| ---- | ---- |
| `(expression...)`, `[expressions...]`, `{expressions...]`, `{key: value...}` | binding or parenthesized expression, list display, set display, dict display |
| `x[index]`, `x[index:index]`, `x(arguments...)`, `x.attribute` | subscription, slicing, call, attribute reference |
| `await x` | await expression |
| `**` | exponentiation |
| `+x`, `-x`, `~x` | positive, negative, bitwise NOT |
| `*`, `@`, `/`, `//`, `%` | multiplication, matrix multiplication, division, floor division, remainder |
| `+`, `-` | addition & subtraction |
| `<<`, `>>` | shifts |
| `&` | bitwise AND |
| `^` | bitwise XOR |
| `|` | bitwise OR |
| `in`, `not in`, `is`, `is not`, `<`, `<=`, `>`, `>=`, `!=`, `==` | Boolean: membership test, comparison tests
| `not x` | Boolean NOT |
| `and` | Boolean AND |
| `or` | Booean OR |
| `if ... else ...` | conditional |
| `lambda` | Lambda expression |
| `:=` | assignment expression |

### Strings

- Strings are ordered collections (sequences) of Unicode characters
- Strings are enclosed in matching quotes, either single (`'`) or double (`"`).

Note:
- Strings are Immutable.
- The `+` operator concatenate strings to form a new string.
- Strings are iterable.
- `len()` function returns length of the string.

#### Examples

````python
name = 'IT 111 Fundamentals of Programming'

print(type(name))

print(len(name))
````

### Option / Condition

```python
if option1:

    ...
    ...
else:
    print('option 3.')
```

**In Python indents are required, not optional.**
  - convention is 4 spaces per indent level.
  - careful not to mix tabs with spaces!
  - set editors to replace tab characters with 4 spaces.
  - usually a problem from sharing code with different indent settings.
    

````python
a = 2

if a < 0:
    print('a is negative')
elif a < 5:
    print('a is less than 5.')
elif a == 5:
    print('a is ten.')
else:
    print('a is greater than 5.')
````
