## 1. Python Syntax

Python uses indentation instead of braces or keywords to define code blocks. Statements end with a newline, and Python is case-sensitive.


```python
temperature = 25
if temperature > 30:
    print("It's hot today!")
else:
    print("The weather is pleasant.")














```

This Python program checks if a number is even and prints "Even" if it is.


```python

number = 8
if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

## 2. Python Comments

Comments help explain your code and are ignored by Python.
- Single-line: `#`
- Multi-line: `""" """`


```python
print("Python is fun!")  # This prints a message
""""
    This is a multi-line comment
    Explaining the code in detail
"""
```

## 3. Python Variables

Variables store values that can be reused and modified.
- Dynamic typing: no need to declare type.
- Names must start with a letter or underscore, no spaces.


```python

city = "Paris"
population_million = 2.1
is_capital = True
print(city, population_million, is_capital)
```

## 4. Python Data Types

- `int` – whole numbers
- `float` – decimals
- `str` – text
- `bool` – True/False
- `list` – ordered collection


```python
#int
x = 42

#float
pi = 3.1416

#str
name = "John"

#bool
active = True

print(x, pi, name, active)
```

## 5. Python User Input

Use `input()` to get user input. Input is always a string, so convert to numbers if needed.


```python

height = float(input("Enter your height in meters: "))
print(f"Your height is {height} meters.")
```

## 6. Python String Formatting

Ways to format strings:
- Concatenate: `"Hello " + name`
- f-strings: `f"Hello {name}"`
- `format()` method: `"Hello {}".format(name)`


```python

first_name = "Liam"
last_name = "Smith"
print("Hello, " + first_name + " " , last_name)
print(f"Hello, {first_name} {last_name}!")
print("Hello, {} {}!".format(first_name, last_name))
```
